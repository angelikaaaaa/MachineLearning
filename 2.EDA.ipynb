```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Implementacja bazy
df_transaction = pd.read_csv('/content/drive/My Drive/train_transaction.csv')
df_identity = pd.read_csv('/content/drive/My Drive/train_identity.csv')
#Połączenie bazy względem TransactionID
df = df_identity.merge(df_transaction, on=['TransactionID'])
```

```
round(df.describe(), 2)
```

```
missingi = df.isnull().sum()
wszystkie_dane = np.product(df.shape)
wszystkie_missingi = missingi.sum()
print ("Procent missingów w bazie ",(wszystkie_missingi/wszystkie_dane) * 100)
```

```
fraud = df[df["isFraud"] == 1]
nonfraud = df[df["isFraud"] == 0]
```

```
df['isFraud'].value_counts()
```

```
print('procent fraudów: ',df['isFraud'].value_counts()[1]/len(df['isFraud'])*100)
```

```
print("Średnia fraudów: \n",fraud.TransactionAmt.mean())
print("Średnia niefraudów: \n", nonfraud.TransactionAmt.mean())
print("Maximum fraudów: \n",fraud.TransactionAmt.max())
print("Minimum fraudów: \n",fraud.TransactionAmt.min())
print("Maximum niefraudów: \n", nonfraud.TransactionAmt.max())
print("Minimum niefraudów: \n", nonfraud.TransactionAmt.min())
```

```
dyskretne = ['id_12', 'id_16', 'id_28', 'id_29',	'id_30', 'id_31', 'id_33', 'id_34', 'id_35',	'id_36',	'id_37',	'id_38',	'DeviceType',	'DeviceInfo',	'ProductCD', 'card4', 'card6', 'P_emaildomain',	'R_emaildomain']
```

```
 for zm_dyskretna in dyskretne:
  f, ax = plt.subplots(1, 2, figsize=(10, 5))
  colors=['red', 'orange', 'yellow', 'green', 'blue', 'pink']
  (df[zm_dyskretna].value_counts().head(6)).plot(kind='bar', title=zm_dyskretna,  ax=ax[0], color=colors)
  ((fraud[zm_dyskretna].value_counts()*100/df[zm_dyskretna].count()).head(6)).plot(kind='barh', title='Procent fraudow', ax=ax[1],color=colors)

  plt.show()
```

```

plt.hist(nonfraud.hours,np.linspace(0,24,25),alpha=.5,density = True,label='Niefraud')
plt.hist(fraud.hours,np.linspace(0,24,25),alpha=.5,density = True,label='Fraud')
plt.legend(loc = 'lower right')
plt.title('Transakcje na godzine')
plt.xlabel('Godzina')
plt.ylabel('Transakcje')
plt.show()
```
